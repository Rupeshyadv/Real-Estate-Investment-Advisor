```python
# import libraries
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

```python
# load models
cls_model = joblib.load('../models/trained_models/xgb_classifier_model.joblib')
reg_model = joblib.load('../models/trained_models/xgb_regressor_model.joblib')
# load features
features = joblib.load("../models/features/features.joblib")
```

```python
# load dataset
df = pd.read_csv('../data/classification_data/india_housing_prices_classification.csv')
df.head()
```

```python
# CLASSIFICATION MODEL EVALUATION
Xcls = df[features]
ycls = (df["Price_per_SqFt"] > df["Price_per_SqFt"].median()).astype(int)

y_pred_cls = cls_model.predict(Xcls)

print("Accuracy:", accuracy_score(ycls, y_pred_cls))
print("F1 Score:", f1_score(ycls, y_pred_cls))
print("Confusion Matrix:\n", confusion_matrix(ycls, y_pred_cls))
```

```python
# REGRESSION MODEL EVALUATION
Xreg = df.drop(columns=["Price_per_SqFt", "Price_in_Lakhs", "Good_Investment"])
yreg = df["Price_per_SqFt"]

y_reg_pred = reg_model.predict(Xreg)

print("Mean Squared Error:", mean_squared_error(yreg, y_reg_pred))
print("R^2 Score:", r2_score(yreg, y_reg_pred))
```
